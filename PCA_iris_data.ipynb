```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
sns.set_theme(style='darkgrid')
plt.rcParams['figure.figsize'] = (10,5)
plt.rcParams['figure.dpi'] = 200
%matplotlib inline
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/aishwaryamate/Datasets/main/Iris.csv')
df
```

```python
df.drop(columns=['Id'], inplace=True)
```

```python
df
```

# Define X and Y

```python
x = df.iloc[:,:-1]
y = df['Species']
```

```python
x
```

# PCA

```python
sc = StandardScaler()

x = sc.fit_transform(x)
x
```

```python
pca = PCA()
pcs = pca.fit_transform(x)
```

```python
pcs
```

```python
pd.DataFrame(pcs)
```

```python
pca.explained_variance_ratio_
```

```python
np.cumsum(pca.explained_variance_ratio_)
```

```python
# plt.figure(figsize=(5,3),dpi=200)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Principal Components')
plt.ylabel('Explained Variance')
plt.title('Scree Plot', size = 50)
```

# With 2 pcs

```python
pca = PCA(n_components = 2)
pcs = pca.fit_transform(x)
pcs
```

```python
data = pd.DataFrame(pcs, columns=['PC1','PC2'])
data
```

```python
sns.scatterplot(x = data['PC1'], y = data['PC2'],s = 50, edgecolor = 'black', hue = y )
```

```python
data['Class'] = y
data
```
